```python
import requests
import json

req = requests.get("https://apis.zigbang.com/v2/items/34645384")

data= json.loads(req.text )
item = data['item']
agent = data['agent']

deposit = item['보증금액']
rent = item['월세금액']
#premium = item['권리금액']

city1 = item['local1']
city2 = item['local2']
city3 = item['local3']

images = item['images']
phone = agent['agent_phone']

data_keys = list(data['item'].keys())
data_values = list(data['item'].values())
columns_list = list(data['item'].keys())
```

```python
import pandas as pd

df = pd.DataFrame()

for i in range(33500000, 35000000):  # 34463000~34464000
    
    #print(i)      
    
    try:
          
        url = "https://apis.zigbang.com/v2/items/" + str(i)
        # print(url)
        req = requests.get(url)
        data= json.loads(req.text)
        item= data['item']
        
        # print(item['local1'])

        if item['local1'] == '서울특별시':
            
            if item['service_type'].find('원룸'):
                
                #print('원룸 발견')        
                
                s = pd.Series(item.values())
                
                df = df.append(s, ignore_index=True)
            
            elif item['service_type'].find('오피스텔'):
                
                #print('오피스텔 발견')        
                
                s = pd.Series(item.values())
                
                df = df.append(s, ignore_index=True)				
            
        else:
            
            #print("예외처리")
            
            pass
                          
    # 나머지 오류 케이스들은 pass 로 처리 
    except:
        
        pass
        
 
df.columns = list(data_keys)

df.to_csv('직방크롤링원룸오피스텔.csv', index= False)
```
