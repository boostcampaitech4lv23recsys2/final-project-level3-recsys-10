```python
# basic_data['residence_type'].nunique() # 181
# basic_data['sales_type'].nunique() # 2
# basic_data['local2'].nunique() # 25
# basic_data['local3'].nunique() # 237

array1, _ = pd.factorize(basic_data['residence_type'])
array2, _ = pd.factorize(basic_data['sales_type'])
array3, _ = pd.factorize(basic_data['local2'])
array4, _ = pd.factorize(basic_data['local3'])

basic_data['residence_type'] = array1
basic_data['sales_type'] = array2
basic_data['local2'] = array3
basic_data['local3'] = array4
```

```python
timestamp = range(1000,1050)
# timestamp = time.monotonic()
interest = 1
user_id = [random.randrange(1, 20, 1) for i in range(50)]
item_id = [random.choice(list(basic_data['item_id'].unique())) for _ in range(50)]

interaction = pd.DataFrame(user_id, columns=['user_id'])
interaction['item_id'] = item_id
interaction['interest'] = interest
interaction['timestamp'] = timestamp

interaction
```

```python
interaction['user_id'].value_counts()
```

```python
interaction.columns = ['user_id:token', 'item_id:token', 'interest:float', 'timestamp:float']
```

### Save
