```python
t1 = theater[theater['영업상태코드']==1]
```

```python
t1 = t1[['관리번호','사업장명','인허가일자','최종수정일자','영업상태명','전화번호','도로명주소','도로명우편번호']][~t1['도로명주소'].duplicated()]
```

```python
tmp = t1
```

```python
def address_preprocessing(x):
    s = str(x).split()
    for i in range(len(s)):
        if s[i][-1]=='로' or s[i][-1]=='길':
            return ' '.join(s[:i+2])
    else:
        return 0
```

```python
tmp['도로명주소'] = tmp['도로명주소'].apply(lambda x: address_preprocessing(x))
# tmp['도로명주소'] = tmp['도로명주소'].apply(lambda x: x.split('(')[0])
```

```python
tmp.shape
```

```python
tmp.drop(147,axis=0,inplace=True)
```

```python
tmp = tmp[~tmp['도로명주소'].duplicated()]
```

```python
pd.options.display.max_rows = 110
```

```python
# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
tmp['위경도'] = tmp['도로명주소'].apply(lambda x: geocoding(x))
```

```python
for index, row in tmp.iterrows():
    tmp.loc[index,'위도'] = row['위경도'][0]
    tmp.loc[index,'경도'] = row['위경도'][1]
```
