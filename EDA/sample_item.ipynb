```python
import pandas as pd
import numpy as np

import random
from tqdm import tqdm
```

```python
data_path = '../data/'
recbole_path = '../modeling/RecBole/dataset/sample/'
```

```python
basic_data = pd.read_csv(data_path+'basic_data_1.1.csv')
basic_data
```

```python
item = basic_data['item_id'].unique()
for i in tqdm(item):
    basic_data.loc[basic_data[basic_data['item_id']==i].index, 'infra_1'] = random.randrange(0, 100, 10)
    basic_data.loc[basic_data[basic_data['item_id']==i].index, 'infra_2'] = random.randrange(0, 100, 5)
    basic_data.loc[basic_data[basic_data['item_id']==i].index, 'infra_3'] = random.randrange(0, 100, 20)
    basic_data.loc[basic_data[basic_data['item_id']==i].index, 'infra_4'] = random.randrange(0, 100, 5)
    basic_data.loc[basic_data[basic_data['item_id']==i].index, 'infra_5'] = random.randrange(0, 100, 10)
```

```python
basic_data
```

```python
# basic_data['residence_type'].nunique() # 181
# basic_data['sales_type'].nunique() # 2
# basic_data['local2'].nunique() # 25
# basic_data['local3'].nunique() # 237

array1, _ = pd.factorize(basic_data['residence_type'])
array2, _ = pd.factorize(basic_data['sales_type'])
array3, _ = pd.factorize(basic_data['local2'])
array4, _ = pd.factorize(basic_data['local3'])

basic_data['residence_type'] = array1
basic_data['sales_type'] = array2
basic_data['local2'] = array3
basic_data['local3'] = array4
```

```python
basic_data
```

```python
basic_data['local2'].value_counts()
```

```python
basic_data[basic_data['local2']==0]
```

```python
item_id = basic_data['item_id'].unique()
basic_data['latitude'] = 0
basic_data['longitude'] = 0
for id in tqdm(item_id):
    loc = ''.join(basic_data[basic_data['item_id']==id]['random_location']).split(',')
    latitude, longitude = loc[0], loc[1]

    basic_data.loc[basic_data[basic_data['item_id']==id].index, 'latitude'] = latitude
    basic_data.loc[basic_data[basic_data['item_id']==id].index, 'longitude'] = longitude
```

```python
basic_data.columns
```

```python
basic_data.drop(columns='random_location',axis=1,inplace=True)
basic_data = basic_data[['item_id', 'residence_type', 'sales_type', 'deposit', 'monthly_rent',
       'exclusive_area', 'supply_area', 'latitude',
       'longitude', 'local2', 'local3',
       'infra_1', 'infra_2', 'infra_3', 'infra_4', 'infra_5']]
```

```python
basic_data.columns = ['item_id:token', 'residence_type:token', 'sales_type:token', 'deposit:float', 'monthly_rent:float',
       'exclusive_area:float', 'supply_area:float', 'latitude:float','longitude:float', 'local2:token', 'local3:token',
       'infra_1:float', 'infra_2:float', 'infra_3:float', 'infra_4:float', 'infra_5:float']
```

### Save

```python
basic_data.to_csv(data_path+'sample_item_data.csv',index=False)
```
